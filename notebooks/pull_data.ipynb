# Data Pull: GoVocal + Typeform

This notebook pulls all data from **GoVocal** (users, ideas, comments, reactions) and **Typeform** (form responses), then saves everything as JSON files in the `data/` folder.

**Prerequisites:**
1. Copy `.example.env` to `.env` and fill in your API credentials
2. Run `pip install -r requirements.txt` in your terminal

**How to use:** Run each cell in order (Shift+Enter). The notebook will show progress bars and previews as data is pulled.

## 1. Setup

```python
import sys, os

# Make sure we can import from the src/ directory
project_root = os.path.abspath(os.path.join(os.getcwd(), ".."))
if project_root not in sys.path:
    sys.path.insert(0, project_root)

import pandas as pd
from src import config
from src.govocal_client import GoVocalClient
from src.typeform_client import TypeformClient

# Validate credentials are set
config.validate()

# Ensure data output directory exists
DATA_DIR = os.path.join(project_root, "data")
os.makedirs(DATA_DIR, exist_ok=True)

print("All credentials loaded successfully.")
print(f"Output directory: {DATA_DIR}")
```

## 2. Configuration

Edit the list below to specify which Typeform forms to pull responses from.
You can find form IDs in the Typeform URL: `https://admin.typeform.com/form/<FORM_ID>/...`

```python
# ============================================================
# EDIT THIS: Add your Typeform form IDs here
# ============================================================
TYPEFORM_FORM_IDS = [
    "KdHzkJeL",  #  SC Forum Final
    "PmPIQkd8",  # SC Forum - PD 
    "YcnYy8ah"  # SC Forum - BA
]
# Include partial (incomplete) Typeform responses?
INCLUDE_PARTIAL_RESPONSES = True

print(f"Will pull {len(TYPEFORM_FORM_IDS)} Typeform form(s)")
print(f"Include partial responses: {INCLUDE_PARTIAL_RESPONSES}")
```

## 3. Pull GoVocal Data

This pulls **all** users, ideas, comments, and reactions from your GoVocal platform. Progress bars will show how many pages are being fetched.

```python
gv = GoVocalClient()
gv.authenticate()
print("Authenticated with GoVocal.\n")

# Pull all data
gv_users = gv.get_users()
gv_ideas = gv.get_ideas()
gv_comments = gv.get_comments()
gv_reactions = gv.get_reactions()

# Convert to DataFrames
df_users = pd.DataFrame(gv_users)
df_ideas = pd.DataFrame(gv_ideas)
df_comments = pd.DataFrame(gv_comments)
df_reactions = pd.DataFrame(gv_reactions)

print(f"\n--- GoVocal Summary ---")
print(f"Users:     {len(df_users):,}")
print(f"Ideas:     {len(df_ideas):,}")
print(f"Comments:  {len(df_comments):,}")
print(f"Reactions: {len(df_reactions):,}")
```

```python
# Preview: Users
print(f"Users columns: {list(df_users.columns)}")
df_users.head()
```

```python
# Preview: Ideas
print(f"Ideas columns: {list(df_ideas.columns)}")
df_ideas.head()
```

## 4. Pull Typeform Data

This pulls all responses (completed + partial if enabled) from the Typeform forms listed in the Configuration cell above.

```python
tf = TypeformClient()

tf_dataframes = {}

if not TYPEFORM_FORM_IDS:
    print("No Typeform form IDs configured. Add form IDs in the Configuration cell above.")
else:
    all_responses = tf.get_all_responses(TYPEFORM_FORM_IDS, include_partial=INCLUDE_PARTIAL_RESPONSES)

    for form_id, responses in all_responses.items():
        rows = [TypeformClient.flatten_response(r, form_id=form_id) for r in responses]
        df = pd.DataFrame(rows)
        tf_dataframes[form_id] = df
        print(f"\nForm {form_id}: {len(df):,} responses, {len(df.columns)} columns")

    print(f"\n--- Typeform Summary ---")
    print(f"Forms pulled: {len(tf_dataframes)}")
    print(f"Total responses: {sum(len(df) for df in tf_dataframes.values()):,}")
```

```python
# Preview: First Typeform (if any)
if tf_dataframes:
    first_id = list(tf_dataframes.keys())[0]
    print(f"Preview of form {first_id}:")
    display(tf_dataframes[first_id].head())
else:
    print("No Typeform data to preview.")
```

## 5. Save to JSON

All data is saved to the `data/` folder as JSON files.

```python
# Save GoVocal data
df_users.to_json(os.path.join(DATA_DIR, "govocal_users.json"), orient="records", indent=2)
df_ideas.to_json(os.path.join(DATA_DIR, "govocal_ideas.json"), orient="records", indent=2)
df_comments.to_json(os.path.join(DATA_DIR, "govocal_comments.json"), orient="records", indent=2)
df_reactions.to_json(os.path.join(DATA_DIR, "govocal_reactions.json"), orient="records", indent=2)
print("Saved GoVocal JSON files:")
print(f"  data/govocal_users.json      ({len(df_users):,} rows)")
print(f"  data/govocal_ideas.json      ({len(df_ideas):,} rows)")
print(f"  data/govocal_comments.json   ({len(df_comments):,} rows)")
print(f"  data/govocal_reactions.json  ({len(df_reactions):,} rows)")

# Save Typeform data
for form_id, df in tf_dataframes.items():
    filename = f"typeform_{form_id}.json"
    df.to_json(os.path.join(DATA_DIR, filename), orient="records", indent=2)
    print(f"  data/{filename}  ({len(df):,} rows)")

print("\nDone! All data saved.")
```

## 6. Summary

Quick overview of everything that was pulled.

```python
summary = {
    "Source": [],
    "Dataset": [],
    "Records": [],
    "File": [],
}

for name, df, fname in [
    ("Users", df_users, "govocal_users.json"),
    ("Ideas", df_ideas, "govocal_ideas.json"),
    ("Comments", df_comments, "govocal_comments.json"),
    ("Reactions", df_reactions, "govocal_reactions.json"),
]:
    summary["Source"].append("GoVocal")
    summary["Dataset"].append(name)
    summary["Records"].append(len(df))
    summary["File"].append(fname)

for form_id, df in tf_dataframes.items():
    summary["Source"].append("Typeform")
    summary["Dataset"].append(f"Form {form_id}")
    summary["Records"].append(len(df))
    summary["File"].append(f"typeform_{form_id}.json")

df_summary = pd.DataFrame(summary)
print(f"Total records across all sources: {df_summary['Records'].sum():,}\n")
df_summary
```
